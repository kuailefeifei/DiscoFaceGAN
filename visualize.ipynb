```python
import os

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_func(image_arrays, num=5):
    n = len(image_arrays)
    num_row = n // num
    plt.figure(figsize=(25, 1.5*num))
    for i, image_array in enumerate(image_arrays):
        plt.subplot(num_row, num, i + 1)
        plt.imshow(image_array)
        plt.axis('off')
    plt.show()
```

```python
def read_images(dataset_path):
    image_list = []
    for image_name in sorted(os.listdir(dataset_path)):
        image_path = os.path.join(dataset_path, image_name)
        image_list.append(image_path)
    return image_list
```

```python
dataset_path = 'generate_images'
image_list = read_images(dataset_path)
image_arrays = []
for image_path in image_list:
    image_array = np.array(Image.open(image_path))
    image_arrays.append(image_array)
plot_func(image_arrays)
```
